```python
import cv2
import numpy as np
import pylab as pl

if __name__ == '__main__':
    # 源读取图像
    im_src = cv2.imread('img_src2.jpg')
    # 定义图像中的锚点，一般是角点
    pts_src = np.array([[241, 287], [165, 360], [295, 450], [361, 364]])

    # 读取目标图像
    im_dst = cv2.imread('img_dst2.jpg')
    # 定义图像中的锚点，一般是角点
    pts_dst = np.array([[160, 237], [142, 377], [365, 392], [365, 253]])

    # 计算单应性矩阵
    h, status = cv2.findHomography(pts_src, pts_dst)

    # 应用单应性矩阵
    im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1], im_dst.shape[0]))

    pl.figure(), pl.imshow(im_src[:, :, ::-1]), pl.title('src'),
    pl.figure(), pl.imshow(im_dst[:, :, ::-1]), pl.title('dst')
    # 展示结果
    pl.figure(), pl.imshow(im_out[:, :, ::-1]), pl.title('out'), pl.show()

    # save image
    cv2.imwrite('out_img2.jpg', im_out)
```
